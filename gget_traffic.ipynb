# Load data

```python
import glob
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
# Get filenames of old and new traffic data
traffic_files = glob.glob("./gget/traffic/*.csv")
```

```python
# Build data frame containing all traffic data
for i, file in enumerate(traffic_files):
  if i == 0:
    # Initiate dataframe
    traffic_df = pd.read_csv(file)
    traffic_df = traffic_df.set_index("_date")

  else:
    temp_df = pd.read_csv(file)
    temp_df = temp_df.set_index("_date")
    # Add new data to data frame
    traffic_df = temp_df.combine_first(traffic_df)

# Drop eventual duplicates
traffic_df = traffic_df.drop_duplicates().fillna(0)

traffic_df.head()
```

# Plotting

```python
import matplotlib.pyplot as plt
%config InlineBackend.figure_format='retina'
```

```python
!mkdir plots
```

```python
fig, axs = plt.subplots(figsize=(10,10), nrows=4, sharex=True)

fontsize=15
lw = 1

color1 = "tab:blue"
color2 = "tab:orange"

x = pd.to_datetime(traffic_df.index.values)

## Plot number of clones
# Total clones
ax = axs[0]
y = traffic_df["total_clones"].values
ax.plot(x, y, label="Total", lw=lw, c=color1)
ax.set_title("$\it{gget}$ Repository Clones Over Time (Total)", fontsize=fontsize)

# Unique clones
ax = axs[1]
x = pd.to_datetime(traffic_df.index.values)
y = traffic_df["unique_clones"].values
ax.plot(x, y, lw=lw, c=color1)
ax.set_title("$\it{gget}$ Repository Clones Over Time (Unique)", fontsize=fontsize)

fig.text(0, 0.75, "# Clones", fontsize=fontsize, ha='center', va='center', rotation='vertical')

## Plot number of views
# Total views
ax = axs[2]
y = traffic_df["total_views"].values
ax.plot(x, y, lw=lw, c=color2, label="Total")
ax.set_title("$\it{gget}$ Repository Views Over Time (Total)", fontsize=fontsize)

# Unique views
ax = axs[3]
y = traffic_df["unique_views"].values
ax.plot(x, y, lw=lw, c=color2)
ax.set_title("$\it{gget}$ Repository Views Over Time (Unique)", fontsize=fontsize)

fig.text(0, 0.3, "# Views", fontsize=fontsize, ha='center', va='center', rotation='vertical')

ax.set_xlabel("Date", fontsize=fontsize)

for ax in axs:
  ax.tick_params(axis='both', labelsize=fontsize-2)

  # Add grid and set below graph
  ax.grid(True, which="both", color="lightgray")
  ax.set_axisbelow(True)

  ax.margins(x=0.01)
  ax.get_yaxis().set_major_formatter(plt.matplotlib.ticker.FuncFormatter(lambda x, _: f'{int(x):,}'))

plt.xticks(rotation=45, ha="right")

plt.tight_layout()

plt.savefig("./plots/gget_clones_views.png", bbox_inches='tight', dpi=300)

fig.show()
```

```python
## Calculate the average number of views and clones per week and per day over the past 6 months
# Get the date range for the last 6 months
traffic_df_copy = traffic_df.copy()
traffic_df_copy.index = pd.to_datetime(traffic_df_copy.index)
end_date = pd.to_datetime(traffic_df_copy.index.values).max()
start_date = end_date - pd.DateOffset(months=6)
filtered_df = traffic_df_copy.loc[start_date:end_date]

# Calculate sums over the last 6 months
total_clones = filtered_df["unique_clones"].sum()
total_views = filtered_df["unique_views"].sum()

# Calculate number of weeks and days
num_days = (end_date - start_date).days
num_weeks = num_days / 7

# Calculate averages
avg_clones_per_day = total_clones / num_days
avg_views_per_day = total_views / num_days
avg_clones_per_week = total_clones / num_weeks
avg_views_per_week = total_views / num_weeks
```

```python
fig, axs = plt.subplots(figsize=(10, 7), nrows=2, sharex=True)

# Add average metrics text
text = (f"6-Month Averages:\n"
        f"• Unique clones: {round(avg_clones_per_week)} per week, {round(avg_clones_per_day)} per day\n"
        f"• Unique views: {round(avg_views_per_week)} per week, {round(avg_views_per_day)} per day")
fig.text(0.5, 0.85, text, fontsize=fontsize - 2, ha='center')

fontsize = 15
lw = 2

color1 = "tab:blue"
color2 = "tab:orange"

x = pd.to_datetime(traffic_df.index.values)

# Calculate cumulative sums
total_clones_cumsum = traffic_df["total_clones"].cumsum().values
unique_clones_cumsum = traffic_df["unique_clones"].cumsum().values
total_views_cumsum = traffic_df["total_views"].cumsum().values
unique_views_cumsum = traffic_df["unique_views"].cumsum().values

# Plot cumulative clones
axs[0].plot(x, total_clones_cumsum, label="Total Clones", lw=lw, c=color1)
axs[0].plot(x, unique_clones_cumsum, label="Unique Clones", lw=lw, c=color1, linestyle='--')

# Plot cumulative views
axs[1].plot(x, total_views_cumsum, label="Total Views", lw=lw, c=color2)
axs[1].plot(x, unique_views_cumsum, label="Unique Views", lw=lw, c=color2, linestyle='--')

# Add labels and legend
axs[0].set_title("$\it{gget}$ GitHub Repository Clones and Views", fontsize=fontsize+2, pad=60)
axs[1].set_xlabel("Date", fontsize=fontsize)
for ax in axs:
  leg = ax.legend(fontsize=fontsize - 2)
  for lh in leg.legend_handles:
      lh.set_alpha(1)

fig.text(-0.03, 0.55, "Cumulative Count", fontsize=fontsize, va='center', rotation='vertical')

# Format plot
for ax in axs:
  ax.tick_params(axis='both', labelsize=fontsize - 2)
  ax.grid(True, which="both", color="lightgray")
  ax.set_axisbelow(True)
  ax.margins(x=0.01)
  ax.get_yaxis().set_major_formatter(plt.matplotlib.ticker.FuncFormatter(lambda x, _: f'{int(x):,}'))

plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.savefig("./plots/gget_cumulative_clones_views.png", bbox_inches='tight', dpi=300)
fig.show()
```
